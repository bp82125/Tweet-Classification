```python
merged = pd.merge(processed_hashtags, removed_hashtag, on='id', how='right', suffixes=('_original', '_append'))
```

```python
merged['text_combined'] = merged.apply(
    lambda row: f"{row['text_append']} {row['text_original']}".strip(), axis=1)
```
