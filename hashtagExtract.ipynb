```python
import pandas as pd

import re
data = pd.read_csv("Disaster_Tweets_train.csv")
```

```python
# Function to extract hashtags
def extract_hashtags(text):
    return " ".join(re.findall(r"#\w+", text))

def remove_non_ascii_hashtags(hashtags):
    return " ".join([tag for tag in hashtags.split() if all(ord(c) < 128 for c in tag)])

# Apply the function to the 'text' column
data['hashtags'] = data['text'].apply(extract_hashtags)
data['hashtags'] = data['hashtags'].apply(remove_non_ascii_hashtags)

# Keep only the 'id' and 'hashtags' columns
data_result = data[['id', 'hashtags']]

data_result[data_result['hashtags'] != ""]  # Filtering to show only rows with hashtags
```

```python
data_result
```

```python
result = data_result.set_index('id')['hashtags'].str.split().to_dict()

result = {k: v for k, v in result.items() if len(v) > 0}

result
```

```python
import json

file_path = './hashtags.json'


with open(file_path, 'w') as f:
    json.dump(result, f)
```
