```python
import pandas as pd

data = pd.read_csv("Disaster_Tweets_train.csv")
```

```python
data.info()
```

```python
data['location'] = data['location'].fillna('')
data['keyword'] = data['keyword'].fillna('')

data.info()
```

```python
data.head()
```

```python
import nltk

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('words')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from process_hashtag import extract_hashtags, process_hashtag_string

data['hashtags'] = data['text'].apply(extract_hashtags)
data['hashtags'] = data['hashtags'].apply(process_hashtag_string)

data['text'] = data['text'].str.replace(r'#\w+', '', regex=True)
data['text'] = data['text'] + ' ' + data['hashtags']

data = data.drop(columns=['hashtags'])

data
```

```python
from process_location import process_location

location = pd.read_csv('location.csv')

data['location'] = location['location'].fillna("").apply(process_location)

data["location"].head(50)
```

```python
from process_keyword import process_keyword

data['keyword'] = data['keyword'].apply(process_keyword)

keyword_list = data['keyword'].to_list()
keyword_set = {keyword for keyword in keyword_list if keyword}

keyword_set
```

```python
data['keyword'] = data.apply(
    lambda row: next(
        (keyword for keyword in keyword_set if keyword in row['text']), '')
    if row['keyword'] == '' else row['keyword'], axis=1
)
```

```python
from process_text import process_text

data['text'] = data['text'].apply(process_text)
data = data[data['text'].notna() & (data['text'] != '')]

data.to_csv("processed_data.csv", index=False)
```

```python
import pandas as pd

new_data = pd.read_csv('processed_data.csv')

new_data = new_data.drop(columns=['id', 'keyword', 'location'])

new_data.reset_index()

new_data.head()

new_data.to_csv('final_data.csv', index=False)
```
